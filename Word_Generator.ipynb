```python
text_seq = "I'm gonna make him an offer he can't "
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

class GPT2:
    def __init__(self):
        self.tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
        self.model = GPT2LMHeadModel.from_pretrained("gpt2")

    def predict_next(self, text_seq, k):
        inputs = self.tokenizer(text_seq, return_tensors="pt")
        self.model.eval()

        # Generate k new tokens
        outputs = self.model.generate(
            inputs["input_ids"],
            max_length=inputs["input_ids"].shape[1] + k,
            do_sample=True,       # random sampling
            top_k=50,             # restrict to top 50 candidates
            temperature=0.7       # controls randomness
        )

        return self.tokenizer.decode(outputs[0])
```

```python

gpt2 = GPT2()
print(gpt2.predict_next(text_seq, 5))
```
